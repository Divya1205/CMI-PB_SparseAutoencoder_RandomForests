```python
import requests
from bs4 import BeautifulSoup
import os

# Base URL of the directory
base_url = 'https://www.cmi-pb.org/downloads/cmipb_challenge_datasets/current/'
output_dir = 'C:/Users/divyas/Documents/hackathons/CMI-PB/Data'

# Function to download files from a directory (with recursion for subdirectories)
def download_files_from_directory(directory_url, output_dir):
    response = requests.get(directory_url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all links in the page
    links = soup.find_all('a')

    # Loop over links and process files or subdirectories
    for link in links:
        file_name = link.get('href')

        # Skip parent directories and empty links
        if not file_name or file_name == '../':
            continue

        # Construct the full URL
        full_url = directory_url + file_name

        # Check if it's a subdirectory (ends with '/')
        if file_name.endswith('/'):
            # Recursively download from subdirectory
            subdirectory_path = os.path.join(output_dir, file_name)
            if not os.path.exists(subdirectory_path):
                os.makedirs(subdirectory_path)
            print(f"Navigating into subdirectory: {full_url}")
            download_files_from_directory(full_url, subdirectory_path)
        else:
            # It's a file, download it
            print(f"Downloading {full_url}")
            file_response = requests.get(full_url)
            if file_response.status_code == 200:
                with open(os.path.join(output_dir, file_name), 'wb') as f:
                    f.write(file_response.content)
                print(f"Successfully downloaded {file_name}")
            else:
                print(f"Failed to download {file_name}. HTTP Status: {file_response.status_code}")

# Create output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Start downloading from the base directory
download_files_from_directory(base_url, output_dir)

print("All files downloaded successfully!")
```
