```python
# List of datasets to process
datasets = [
    'pbmc_cell_frequency_normalized_data',
    'pbmc_gene_expression_tpm_raw_data',
    'plasma_ab_titer_normalized_data',
    'plasma_cytokine_concentrations_by_legendplex_normalized_data',
    'plasma_cytokine_concentrations_by_olink_normalized_data',
    't_cell_activation_raw_data',
    't_cell_polarization_raw_data'
]

# Split metadata into training and challenge metadata
# training_meta_data_processed = metadata[metadata['dataset'].str.contains('2020|2021|2022')].copy()
# challenge_meta_data_processed = metadata[metadata['dataset'].str.contains('2023')].copy()

# Ensure specimen_id is the same data type in both datasets and metadata (convert to string)
training_meta_data_processed['specimen_id'] = training_meta_data_processed['specimen_id'].astype(str)
challenge_meta_data_processed['specimen_id'] = challenge_meta_data_processed['specimen_id'].astype(str)

# Separate and transpose datasets for training and challenge, and merge metadata
training_datasets = []
challenge_datasets = []

for dataset_name in datasets:
    dataset = globals()[dataset_name]
    
    # Transpose the dataset only once
    transposed_dataset = dataset.transpose().copy()
    transposed_dataset.index.name = 'specimen_id'
    transposed_dataset.columns.name = 'feature_name'
    
    # Convert specimen_id to string in transposed dataset
    transposed_dataset.index = transposed_dataset.index.astype(str)
    
    # Merge metadata for training dataset
    training_dataset = transposed_dataset[transposed_dataset.index.isin(training_meta_data_processed['specimen_id'])]
    training_dataset = training_dataset.merge(training_meta_data_processed.set_index('specimen_id'), left_index=True, right_index=True, how='left')
    
    # Merge metadata for challenge dataset
    challenge_dataset = transposed_dataset[transposed_dataset.index.isin(challenge_meta_data_processed['specimen_id'])]
    challenge_dataset = challenge_dataset.merge(challenge_meta_data_processed.set_index('specimen_id'), left_index=True, right_index=True, how='left')
    
    training_datasets.append(training_dataset)
    challenge_datasets.append(challenge_dataset)

# Now we have the transposed and merged training and challenge datasets in their respective lists

# Function to print the number of specimens per subject for both training and challenge datasets together
def print_specimens_per_subject_together(train_datasets, challenge_datasets, train_meta_data, challenge_meta_data, dataset_names):
    for i, dataset_name in enumerate(dataset_names):
        # Process the training dataset
        train_dataset = train_datasets[i].reset_index()  # Already transposed earlier, no need to transpose again
        specimens_per_subject_train = train_dataset.groupby('subject_id')['specimen_id'].nunique()
        train_size = train_datasets[i].shape
        
        # Process the challenge dataset
        challenge_dataset = challenge_datasets[i].reset_index()  # Already transposed earlier, no need to transpose again
        specimens_per_subject_challenge = challenge_dataset.groupby('subject_id')['specimen_id'].nunique()
        challenge_size = challenge_datasets[i].shape
        
        # Print the dataset name, size, and specimens per subject for both training and challenge
        print(f"\n{dataset_name} - Training Dataset - Size: {train_size[0]} rows, {train_size[1]} columns")
        print(f"Training Specimens per subject:\n{specimens_per_subject_train}")
        
        print(f"\n{dataset_name} - Challenge Dataset - Size: {challenge_size[0]} rows, {challenge_size[1]} columns")
        print(f"Challenge Specimens per subject:\n{specimens_per_subject_challenge}")

# function call
print_specimens_per_subject_together(
    training_datasets,
    challenge_datasets,
    training_meta_data_processed,
    challenge_meta_data_processed,
    datasets
)
```

### Task IGg_PT

```python
def prepare_abtiter_data(df_train, df_test):
    # Prepare the data for training based on timepoint
    day_0_data_train = df_train[df_train['timepoint'] == 0].set_index('subject_id')
    day_14_data_train = df_train[df_train['timepoint'] == 14].set_index('subject_id')

    # Filter only those subjects that have both day 0 and day 14 data
    day_0_data_train = day_0_data_train[day_0_data_train.index.isin(day_14_data_train.index)]
    day_14_data_train = day_14_data_train[day_14_data_train.index.isin(day_0_data_train.index)]

    # Retain all features for training, excluding 'IgG_PT' for the features (X) and using it as labels (y)
    X_train = day_0_data_train.drop(columns=['IgG_PT'])
    y_train = day_14_data_train['IgG_PT']

    # Prepare the data for test based on timepoint
    day_0_data_test = df_test[df_test['timepoint'] == 0].set_index('subject_id')
    X_test = day_0_data_test[X_train.columns]  # Ensure test data matches the training data columns
    
    return X_train, y_train, X_test

# function call for plasma_ab_titer_normalized_data
X_train_abtiter, y_train_abtiter, X_test_abtiter = prepare_abtiter_data(training_datasets[2], challenge_datasets[2])
```

## Task Monocytes

```python
def prepare_monocyte_data(df_train, df_test):
    """
    Prepare the data for monocyte prediction based on timepoint.

    Parameters:
    - df_train: Training DataFrame containing all training data.
    - df_test: Test DataFrame containing all challenge/test data.

    Returns:
    - X_train: Features for training (day 0).
    - y_train: Labels for training (day 1 monocyte frequency).
    - X_test: Features for testing (day 0).
    """

    # Prepare the data for training based on timepoint
    day_0_data_train = df_train[df_train['timepoint'] == 0].set_index('subject_id')
    day_1_data_train = df_train[df_train['timepoint'] == 1].set_index('subject_id')

    # Filter only those subjects that have both day 0 and day 1 data
    day_0_data_train = day_0_data_train[day_0_data_train.index.isin(day_1_data_train.index)]
    day_1_data_train = day_1_data_train[day_1_data_train.index.isin(day_0_data_train.index)]

    # Retain all features for training, excluding 'Monocytes' for features (X) and using it as labels (y)
    X_train = day_0_data_train.drop(columns=['Monocytes'])
    y_train = day_1_data_train['Monocytes']

    # Prepare the data for test based on timepoint
    day_0_data_test = df_test[df_test['timepoint'] == 0].set_index('subject_id')
    X_test = day_0_data_test[X_train.columns]  # Ensure test data matches the training data columns
    
    return X_train, y_train, X_test

# Function call for monocyte data
X_train_monocytes, y_train_monocytes, X_test_monocytes = prepare_monocyte_data(training_datasets[0], challenge_datasets[0])
```
